## O PROBLEMA QUE VAMOS RESOLVER

Imagine que você é um **professor** e quer prever se um aluno vai:
- **PASSAR** (aprovado) → número **1**
- **REPROVAR** → número **0**

Para fazer isso, vamos olhar **3 características** do aluno:

| Característica | Significado | Valores |
|----------------|-------------|----------|
| 1. **Frequência** | Vem às aulas? (≥75%) | 1=SIM, 0=NÃO |
| 2. **Entregas** | Entrega os trabalhos? | 1=SIM, 0=NÃO |
| 3. **Nota** | Tira nota boa? (≥6.0) | 1=SIM, 0=NÃO |

### Como representar um aluno:

Cada aluno vira uma **lista com 3 números**. Exemplo:

- `[1, 0, 1]` significa:
  - Frequência boa = **1** (SIM)
  - Entrega trabalhos = **0** (NÃO)
  - Nota boa = **1** (SIM)

## ETAPA 1: Importar as Ferramentas

Estas são as "ferramentas mágicas" que vamos usar para criar nosso programa inteligente.

**Execute a célula abaixo** (clique nela e aperte `Shift+Enter`):

```python
from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score

print("✓ Ferramentas importadas com sucesso!")
print("Agora temos tudo para criar nosso programa inteligente!")
```

## ETAPA 2 e 3: Criar Exemplos de Alunos APROVADOS

Vamos criar 3 exemplos de alunos que **PASSARAM** (foram aprovados).  
Estes são nossos exemplos para **ENSINAR** o computador.

Veja o padrão: alunos aprovados geralmente têm **pelo menos 2 coisas boas**!

```python
# Aluno aprovado 1: tem tudo! (frequência, entregas, nota)
aprovado1 = [1, 1, 1]

# Aluno aprovado 2: frequência boa, entregas OK, mas nota não tão boa
aprovado2 = [1, 1, 0]

# Aluno aprovado 3: frequência boa, não entregou tudo, mas nota boa
aprovado3 = [1, 0, 1]

print("✓ Alunos aprovados criados:")
print(f"  Aprovado 1: {aprovado1}")
print(f"  Aprovado 2: {aprovado2}")
print(f"  Aprovado 3: {aprovado3}")
```

## ETAPA 4: Criar Exemplos de Alunos REPROVADOS

Agora vamos criar 3 exemplos de alunos que **REPROVARAM**.

Veja o padrão: alunos reprovados têm **muita coisa faltando**!

```python
# Aluno reprovado 1: não tem nada (sem frequência, sem entregas, sem nota)
reprovado1 = [0, 0, 0]

# Aluno reprovado 2: só entregou trabalhos, mas faltou e nota ruim
reprovado2 = [0, 1, 0]

# Aluno reprovado 3: faltou muito, não entregou, só a nota foi OK
reprovado3 = [0, 0, 1]

print("✓ Alunos reprovados criados:")
print(f"  Reprovado 1: {reprovado1}")
print(f"  Reprovado 2: {reprovado2}")
print(f"  Reprovado 3: {reprovado3}")
```

## ETAPA 5 e 6: Juntar Todos os Dados

Agora vamos:
1. Juntar TODOS os alunos em uma lista (chamada de "dados de treino")
2. Criar a lista de respostas corretas (o "gabarito" de treino)

```python
# Juntar todos os alunos
dados = [aprovado1, aprovado2, aprovado3, reprovado1, reprovado2, reprovado3]

# Criar o gabarito (1=APROVADO, 0=REPROVADO)
# Os 3 primeiros são aprovados: 1, 1, 1
# Os 3 últimos são reprovados: 0, 0, 0
classes = [1, 1, 1, 0, 0, 0]

print("="*50)
print("DADOS DE TREINO (características dos alunos):")
print("="*50)
print(dados)
print()
print("RESPOSTAS CORRETAS (gabarito):")
print(classes)
print()
print("Explicando melhor:")
for i in range(len(dados)):
    if classes[i] == 1:
        resultado = "APROVADO ✓"
    else:
        resultado = "REPROVADO ✗"
    print(f"Aluno {i+1}: {dados[i]} → {resultado}")
```

## ETAPA 7: TREINAR O MODELO (A Parte Mágica!)

Agora vem a parte mais legal!

Vamos criar um **modelo** (programa inteligente) e **ENSINAR** ele a reconhecer padrões.

O comando `fit()` significa "treinar" em inglês.

```python
# Criar o modelo
modelo = LinearSVC(max_iter=10000)

# TREINAR o modelo (ensinar os padrões)
modelo.fit(dados, classes)

print("🎓 MODELO TREINADO COM SUCESSO!")
print("Agora ele aprendeu os padrões dos alunos aprovados e reprovados!")
```

## ETAPA 8: Criar Alunos para Testar

Agora vamos criar alunos que o modelo **NUNCA VIU** antes.

Vamos ver se ele consegue **ADIVINHAR** se eles vão passar ou reprovar!

**Pense você também:** o que VOCÊ acha de cada aluno?

```python
# Aluno misterioso 1: tem TUDO (frequência, entregas, nota)
# O que você acha? Vai passar ou reprovar?
misterioso1 = [1, 1, 1]

# Aluno misterioso 2: tem frequência e entregas, mas nota ruim
# O que você acha? Vai passar ou reprovar?
misterioso2 = [1, 1, 0]

# Aluno misterioso 3: falta muito, não entrega, mas nota boa
# O que você acha? Vai passar ou reprovar?
misterioso3 = [0, 0, 1]

# Juntar os 3 alunos misteriosos
alunos_teste = [misterioso1, misterioso2, misterioso3]

print("="*50)
print("ALUNOS PARA TESTAR O MODELO:")
print("="*50)
print("Aluno 1:", misterioso1, "→ [Freq=SIM, Entregas=SIM, Nota=SIM]")
print("Aluno 2:", misterioso2, "→ [Freq=SIM, Entregas=SIM, Nota=NÃO]")
print("Aluno 3:", misterioso3, "→ [Freq=NÃO, Entregas=NÃO, Nota=SIM]")
```

## ETAPA 9: Fazer as Previsões

Agora vamos pedir para o modelo **PREVER** (fazer um palpite).

O comando `predict()` significa "prever" em inglês.

```python
# Pedir para o modelo fazer as previsões
previsoes = modelo.predict(alunos_teste)

# Mostrar as previsões
print("="*50)
print("PREVISÕES DO MODELO:")
print("="*50)
for i in range(len(alunos_teste)):
    if previsoes[i] == 1:
        resultado = "APROVADO ✓"
    else:
        resultado = "REPROVADO ✗"
    print(f"Aluno {i+1}: {alunos_teste[i]} → Previsão: {resultado}")
```

## ETAPA 10: CRIAR O GABARITO (Aqui você precisa pensar!)

**ATENÇÃO:** Esta é a parte que VOCÊ precisa preencher!

Pense bem em cada aluno e decida:
- Aluno 1 `[1,1,1]`: tem tudo! Deveria passar ou reprovar?
- Aluno 2 `[1,1,0]`: frequência boa, entregas OK, nota ruim. E aí?
- Aluno 3 `[0,0,1]`: só a nota foi boa, resto tudo ruim. E aí?

**Substitua os `?` por:**
- `1` se você acha que deve ser **APROVADO**
- `0` se você acha que deve ser **REPROVADO**

```python
# PREENCHA AQUI - Substitua os ? por 1 ou 0
# Exemplo: gabarito = [1, 1, 0] significa:
#   - Aluno 1: APROVADO (1)
#   - Aluno 2: APROVADO (1)
#   - Aluno 3: REPROVADO (0)

gabarito = [?, ?, ?]  # ← PREENCHA AQUI!

print("✓ Seu gabarito:", gabarito)
```

## ETAPA 11: Comparar Previsões com o Gabarito

Agora vamos ver quantas vezes o modelo **ACERTOU** comparado com o que VOCÊ acha que é certo!

```python
print("="*50)
print("COMPARANDO PREVISÕES COM SEU GABARITO:")
print("="*50)

acertos = 0
erros = 0

for i in range(len(alunos_teste)):
    prev_texto = "APROVADO" if previsoes[i] == 1 else "REPROVADO"
    gab_texto = "APROVADO" if gabarito[i] == 1 else "REPROVADO"

    if previsoes[i] == gabarito[i]:
        status = "✓ ACERTOU!"
        acertos = acertos + 1
    else:
        status = "✗ ERROU!"
        erros = erros + 1

    print(f"\nAluno {i+1}:")
    print(f"  O modelo previu: {prev_texto}")
    print(f"  Você acha que é: {gab_texto}")
    print(f"  {status}")

print(f"\n{'='*50}")
print(f"RESUMO: {acertos} acertos e {erros} erros")
print(f"{'='*50}")
```

## ETAPA 12: Calcular a Taxa de Acerto

Vamos calcular quantos % o modelo acertou!

```python
# Jeito 1: Calculando "na mão"
total_alunos = len(alunos_teste)
taxa_acerto = (acertos / total_alunos) * 100

print("="*50)
print("TAXA DE ACERTO DO MODELO:")
print("="*50)
print(f"O modelo acertou {acertos} de {total_alunos} alunos")
print(f"Taxa de acerto: {taxa_acerto:.1f}%")
print()

# Jeito 2: Usando a ferramenta do sklearn
acuracia = accuracy_score(gabarito, previsoes)
print(f"Confirmando com o sklearn: {acuracia * 100:.1f}%")
```
